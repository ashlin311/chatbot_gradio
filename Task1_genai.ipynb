```python
import os
from google.colab import userdata

os.environ['GROQ_KEY']= userdata.get('GROQ_KEY')
print("Key is set")
```

```python
from groq import Groq

client = Groq(api_key = os.environ['GROQ_KEY'])
```

```python
import gradio as gr

# Initial system prompt, which will also be the default value for the input textbox
INITIAL_SYSTEM_PROMPT = "You are a precise software engineer. Analyze the user request and provide the most efficient solution possible. Your response must consist only of the raw code block and a brief, one-sentence logical explanation underneath."

def chat_function(message, history, temperature, system_prompt_input):
    messages = [{"role": "system", "content": system_prompt_input}]

    for msg in history:
        messages.append({
            "role": msg["role"],
            "content": msg["content"]
        })

    messages.append({"role": "user", "content": message})

    stream = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        temperature=temperature,
        stream=True,
    )

    partial_response = ""
    for chunk in stream:
        delta = chunk.choices[0].delta.content
        if delta:
            partial_response += delta
            yield partial_response

temperature_slider = gr.Slider(
    minimum=0.0,
    maximum=1.0,
    value=0.7,
    step=0.1,
    label="Temperature",
    info="Controls the randomness of the responses"
)

system_prompt_textbox = gr.Textbox(
    value=INITIAL_SYSTEM_PROMPT,
    label="System Prompt",
    info="Define the AI's persona or instructions"
)

demo = gr.ChatInterface(
    chat_function,
    chatbot=gr.Chatbot(height=300, type='messages'),
    textbox=gr.Textbox(placeholder="Ask me a question", container=False, scale=7),
    title="ChatBot with Temperature and System Prompt Control",
    additional_inputs=[
        temperature_slider,
        system_prompt_textbox
    ],
    examples=[["What is the capital of France?"], ["Explain quantum computing simply."]],
)

demo.launch(debug=True)
```
